```python
import cv2
import numpy as np
import imutils
import time


camera = cv2.VideoCapture("visiontraffic.avi")
frameCounter = 0
ret, frame = camera.read()

while (1):
    print "frame:", int(camera.get(cv2.cv.CV_CAP_PROP_FRAME_COUNT)), "ret:", ret
    print "frameCounter:", frameCounter
    if frameCounter == 531:
        print "resetting videostream"
        frameCounter = 0
        camera = cv2.VideoCapture('visiontraffic.avi')
        ret,frame = camera.read()
        print "ret,frame in if statement:", ret, frame
    frameCounter += 1
    frame = imutils.resize(frame, width=500)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow("test",gray)
    key = cv2.waitKey(30) & 0xFF
    if (key == ord('q')):
        break
    ret,frame = camera.read()
    
print 'out of while loop'
camera.release()
cv2.destroyAllWindows()
```
