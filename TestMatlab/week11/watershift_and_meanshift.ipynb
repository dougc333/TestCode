```python
#meanshift
#https://github.com/fjean/pymeanshift
azs
import numpy as np
import cv2
import pymeanshift as pms
import matplotlib.pyplot as plt

original_image = cv2.imread("example-orig.jpg")

(segmented_image, labels_image, number_regions) = pms.segment(original_image, spatial_radius=6, 
                                                              range_radius=4.5, min_density=50)

fig,(ax1,ax2) = plt.subplots(ncols=2, sharex=True, sharey=True, subplot_kw={'adjustable':'box-forced'})

ax1.imshow(original_image)
ax1.set_title('Original')
ax1.axis('off')

ax2.imshow(segmented_image)
ax2.set_title('meanshift')
ax2.axis('off')

plt.show()









```
