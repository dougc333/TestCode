```python
import numpy as np
import scipy.misc

#compute probability model for image 
img = scipy.misc.imread('/Users/dc/TestCode/TestMatlab/week8/Cameraman256.bmp')
img.shape
```

```python
import collections

c = collections.Counter()
for x in np.nditer(img):
    y=int(x)
    c[y] +=1
print c
```

```python
sum(c.values())
```

```python
import math
ent=0
for k,v in c.items():
    c[k] = v/65536.0
    ent = ent+(v/65536.0 * math.log(v/65536.0,2))
print c
print -1*ent
```
