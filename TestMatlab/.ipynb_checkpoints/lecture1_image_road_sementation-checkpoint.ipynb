```python
import numpy as np
import scipy.ndimage as ndi
from skimage import feature
from skimage.filters import roberts, sobel, scharr, prewitt,laplace
import matplotlib.pyplot as plt
import cv2

from skimage.filters import laplace,sobel,roberts
import scipy.ndimage as ndi
from skimage import feature
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import cv2
import pymeanshift as pms
```

```python
%matplotlib inline

im = np.zeros((128, 128))
im[32:-32, 32:-32] = 1

im = ndi.rotate(im, 25, mode='constant')
im = ndi.gaussian_filter(im, 4) 
im += 0.2 * np.random.random(im.shape)


# Compute the Canny filter for two values of sigma
edges1 = feature.canny(im)
edges2 = feature.canny(im, sigma=3)

sobel=sobel(im)
prewitt = prewitt(im)  
roberts = roberts(im)
laplace = laplace(im)

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(8, 3), sharex=True, sharey=True)
fig1, (ax11, ax12, ax13, ax14, ax15) = plt.subplots(nrows=1, ncols=5, figsize=(8, 3), sharex=True, sharey=True)

ax1.imshow(im, cmap=plt.cm.jet)
ax1.axis('off')
ax1.set_title('noisy image', fontsize=20)

ax2.imshow(edges1, cmap=plt.cm.gray)
ax2.axis('off')
ax2.set_title('Canny filter, $\sigma=1$', fontsize=20)

ax3.imshow(edges2, cmap=plt.cm.gray)
ax3.axis('off')
ax3.set_title('Canny filter, $\sigma=3$', fontsize=20)

fig.subplots_adjust(wspace=0.02, hspace=0.02, top=0.9,
                    bottom=0.02, left=0.02, right=0.98)

ax11.imshow(im, cmap=plt.cm.jet)
ax11.axis('off')
ax11.set_title('noisy', fontsize=20)

ax12.imshow(sobel, cmap=plt.cm.gray)
ax12.axis('off')
ax12.set_title('Sobel',  fontsize=20)

ax13.imshow(roberts, cmap=plt.cm.gray)
ax13.axis('off')
ax13.set_title('Prewitt',  fontsize=20)

ax14.imshow(roberts, cmap=plt.cm.gray)
ax14.axis('off')
ax14.set_title('Roberts',  fontsize=20)

ax15.imshow(laplace, cmap=plt.cm.gray)
ax15.axis('off')
ax15.set_title('Laplace',  fontsize=20)

fig1.subplots_adjust(wspace=0.02, hspace=0.10, top=0.9,
                    bottom=0.02, left=0.02, right=0.98)
plt.show()


```

<h4> linear filters can be used to estimate gradients or first derivatives</h4>
Source: http://www.cs.cornell.edu/workshop/curie2015/filtering.pdf

<p>Transform an image to a collection of edges to reduce the amount of information we have to store. 
What is an edge in an image? The area/direction/vector with greatest intensity changes. 
In the first image we have an image with a high white intensity going to a lower black intensity then 
back to a white intensity. The second graph shows the intensity variation and the third graph the derivative 
of the intensity variation. 
</p>
<img src='png_ipython_notebook/image_gradient1.png' style="width: 25%; height: 25%">
<img src='png_ipython_notebook/image_gradient2.png' style="width: 25%; height: 25%">
<img src='png_ipython_notebook/image_gradient3.png' style="width: 25%; height: 25%">

      
We can implement a 1d derivative as a difference equation
We can implement a 2d derivative as a matrix multiplication
    
    

The thick edges are artifacts of threhsholding. When we take a first derivative we get a peak. The width of the peak corresponds to the thickness of the line. 
taken from http://www.owlnet.rice.edu/~elec539/Projects97/morphjrks/laplacian.html

<img src = 'week11/derivative.png'>

<img src = 'week11/derivativepeak.png'>

To reduce the thickness of the lines we can use the second derivative. 
The problem with the second derivative is with false edges generated from artificial 0 crossings 
from pre gaussian noise filterering. 




Add noise adding false derivatives
Add gaussian smoothing removing false derivatives

```python
#example using meanshift for lane segmentation

from skimage.filters import laplace,sobel,roberts
import scipy.ndimage as ndi
from skimage import feature
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import cv2
import pymeanshift as pms


lane1 =  ndi.imread('laneimg1.jpeg')

lane1_opencv = cv2.imread("laneimg1.jpeg")
lane2_opencv = cv2.imread("laneimg2.jpeg")
lane3_opencv = cv2.imread("laneimg3.jpeg")

(meanshift_lane1, labels_image, number_regions) = pms.segment(lane1_opencv, spatial_radius=6,
                                                              range_radius=4.5, min_density=50)

(meanshift_lane2, labels_image, number_regions) = pms.segment(lane2_opencv, spatial_radius=6,
                                                              range_radius=4.5, min_density=50)

(meanshift_lane3, labels_image, number_regions) = pms.segment(lane3_opencv, spatial_radius=6,
                                                              range_radius=4.5, min_density=50)


fig4,(i0,i1) = plt.subplots(nrows=1, ncols=2, figsize=(8, 3), sharex=True, sharey=True)
fig5,(i2,i3) = plt.subplots(nrows=1, ncols=2, figsize=(8, 3), sharex=True, sharey=True)
fig6,(i4,i5) = plt.subplots(nrows=1, ncols=2, figsize=(8, 3), sharex=True, sharey=True)


i0.imshow(lane1)
i0.axis('off')
i0.set_title('orig_lane1',  fontsize=20)

i1.imshow (meanshift_lane1)
i1.axis('off')
i1.set_title('meanshift_lane1',  fontsize=20)
                                                             
i2.imshow(lane2_opencv)
i2.axis('off')
i2.set_title('orig_lane2',  fontsize=20)

i3.imshow (meanshift_lane2)
i3.axis('off')
i3.set_title('meanshift_lane2',  fontsize=20)

i4.imshow(lane3_opencv)
i4.axis('off')
i4.set_title('orig_lane3',  fontsize=20)

i5.imshow (meanshift_lane3)
i5.axis('off')
i5.set_title('meanshift_lane3',  fontsize=20)

                                                              
plt.show()
```
