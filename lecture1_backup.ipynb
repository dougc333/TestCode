<h4>Hough Transform</h4>

HT maps a collection of edge pixels which incompletely represent lines to continuous lines
The points in (x,y) image space are represented as potential lines in the image with the equation y=mx+b
Plot the equations slope/y-intercept in (x,y) space into a m/b Hough space. Each proposed line becomes a point in Hough space. 

There is a duality between a point in (x,y) space <-> a line in Hough Space:
<img src='hough1.png' style="width: 50%; height: 50%">


Given multiple edge points in x,y space which is the result of a line detection algorithm 
we histogram/bin the intersections of the points in Hough space. The intersections of the lines in Hough Space
represent a single line with slope m and y-intercept b in (x,y) space
The bin sizes account for small differences in m/b and instead of having a 
group of small lines together we should have 1 line per bin.


Divide the Hough Space grid and each grid which receives the majority votes become the dominant lines in the image
<img src='hough3.png' style="width: 25%; height: 25%">

We can not represent an vertical slope using cartesion representation so we convert to polar representation. 
This introduces sinusoids into the HT visualizations

<img src='hough4.png' style="width: 25%; height: 25%">

d = y*cos(90-theta)+x*cos(theta) = y(cos(90)cos(theta) + sin(90)sin(theta)) + x*cos(theta)
= y*sin(theta) + x*cos(theta)

A point in (x,y) polar coordinates becomes a sinusoid in Hough space. 
Tuning procedure




```python
#Hough Transform
import skimage
from skimage.filters import laplace,sobel,roberts
import scipy.ndimage as ndi
from skimage import feature
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import cv2
import pymeanshift as pms
import numpy as np
from skimage.transform import (hough_line, hough_line_peaks,
                               probabilistic_hough_line)

#hough examples:
#https://github.com/hmartiro/project-mvq/blob/master/python/notebooks/lane_detection.ipynb
#kalman filter:
#https://github.com/marshallixp/DriverAssist/blob/master/LaneSystem/LaneDetector1.2/TrackLanes.cpp

lane1_opencv = cv2.imread("laneimg1.jpeg")
#lane2_opencv = cv2.imread("laneimg2.jpeg")
#lane3_opencv = cv2.imread("laneimg3.jpeg")

#convert to gray scale from RGB
gray_orig_1 = skimage.color.rgb2gray(lane1_opencv)
#gray_orig_2 = skimage.color.rgb2gray(lane2_opencv)
#gray_orig_3 = skimage.color.rgb2gray(lane3_opencv)


#canny detect
canny_orig_1 = feature.canny(gray_orig_1, sigma=3)
#canny_orig_2 = feature.canny(gray_orig_2, sigma=3)
#canny_orig_3 = feature.canny(gray_orig_3, sigma=3)

r, theta, d = hough_line(canny_orig_1)

fig5, (ax1,ax2) = plt.subplots(1, 2, figsize=(8,4))
fig6, (ax3,ax4) = plt.subplots(1, 2, figsize=(8,4))

ax1.imshow(gray_orig_1,cmap=plt.cm.gray)
ax1.axis('off')
ax1.set_title('orig_lane1',  fontsize=20)

ax2.imshow(canny_orig_1,cmap=plt.cm.gray)
ax2.axis('off')
ax2.set_title('canny_orig_lane1',  fontsize=20)


ax3.imshow(np.log(1 + r),
             extent=[np.rad2deg(theta[-1]), np.rad2deg(theta[0]),
                     d[-1], d[0]],
             cmap=plt.cm.gray, aspect=3.0/5.0)
ax3.set_title('Hough transform')
ax3.set_xlabel('Angles (degrees)')
ax3.set_ylabel('Distance (pixels)')
ax3.axis('image')

ax4.imshow(canny_orig_1,cmap=plt.cm.gray)
ax4.axis('off')
ax4.imshow(gray_orig_1, cmap=plt.cm.gray)
rows, cols = gray_orig_1.shape
for _, angle, dist in zip(*hough_line_peaks(r, theta, d, min_distance=10,min_angle=30,threshold=30.0 ,num_peaks=6)):
    y0 = (dist - 0 * np.cos(angle)) / np.sin(angle)
    y1 = (dist - cols * np.cos(angle)) / np.sin(angle)
    ax4.plot((0, cols), (y0, y1), '-r')
ax4.axis((0, cols, rows, 0))
ax4.set_title('Detected lines')
ax4.set_axis_off()


plt.show()

```
