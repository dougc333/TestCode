```python
import pandas as pd
from collections import Counter,defaultdict
df = pd.read_csv('jacques_wines.tsv')

print(df.head())
print(df.count())
#print(df.tail())
#687 total rows
#for cols in df.columns:
#    print(cols)
print(type(df.columns))
print(df.columns)
print(type(df.index))
print(df.index)
print(type(df.values))
#print(df.values)
list_varieties = []
viney={}
var = defaultdict(int)
for x in df.values:
    vineyard = x[0].split('\t')[0]
    varieties = x[0].split('\t')[1]
    if vineyard not in viney:
        s=set()
        s.add(varieties)
        viney[vineyard] = s
    else:
        viney[vineyard].add(varieties)
    list_varieties.append(x[0].split('\t')[1])
    var[x[0].split('\t')[1]] +=1
print("list_varieties",len(set(list_varieties)))
print(max(var.items(),key=lambda x:x[1]))
print(sorted(var.items(),key=lambda x:x[1]))
#which winery greatest number of varities? 
print('------------------')
print(viney)
print('------------------')
max=0
for k in viney.keys():
    if len(viney[k]) > max:
        max = len(viney[k])
    print("key:",k, len(viney[k]))
    
print('max:',max) #key: Cono Sur 4
#Zinfandel 12th most popular
```

```python
def MinElementsSum(n,S):
  if n==0:
    return 0
  elif n<min(S):
    return None
  cand = []
  for s in S:
    c = MinElementsSum(n-s,S)
    if c is not None:
      cand.append(c)
  if len(cand) == 0:
    return None
  return min(cand)

print(MinElementsSum(2,[1,2]))
0
```
