```python
#5878 with more than 1 entry
c.execute("select count(distinct(code)) from medications group by code having count(code)>1;")
num=c.fetchall()
print len(num)

c.execute("select count(distinct(code)) from medications group by code having count(code)=1;")
num=c.fetchall()
print len(num)

#4057 with one entry, will need more data
c.execute("select code from medications group by code having count(code)>1;")
num=c.fetchall()
print len(num)
#there are 5878 entries with more than 1 code. dictionary/code
#dictTerms={}
numProcessed=0;
writeMe= open("/Users/dc/TestCode/TestCoding/csvfile.csv","w")
for n in num:
    c.execute("select distinct(code_name) from medications where code='%s'" % (n[0]))
    res = c.fetchall()
    #print "res len:",len(res)
    terms=[]
    if len(res) > 1:
        #print "code:%s" % n[0]
        for r in res:
            #print "processing term:%s" % r[0]
            terms.append(r[0])
        writeMe.write(n[0])
        writeMe.write(",")
        for t in terms:
            writeMe.write(t)
            writeMe.write(",")
        writeMe.write("\n")
#        writeMe.write("----------------")
        writeMe.write("\n")
    numProcessed+=1
    if(numProcessed % 100==0):
        print "numProcessed:%s" %numProcessed
writeMe.close()
print "numProcessed:%s" % numProcessed
```
