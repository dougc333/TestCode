```python
import re
import matplotlib.pyplot as plt
from collections import defaultdict

num2=0
numTokens=0
numTerms = defaultdict(int)
terms = defaultdict(int)

for line in open("/Users/dc/TestCode/TestCoding/nomatch.csv"):
    print "line:%s" % line.lower().split("|")[0]
    foo = re.sub(r"[()]","",line.lower().split("|")[0])
    print foo
    # removePO= re.sub(r"po","",foo)
    # print removePO
    print len(foo.split())
    numTerms[len(foo.split())] += 1
    for tokens in foo.split():
        if len(tokens) <=2:
            num2+=1
            if not(tokens.isdigit()):
                terms[tokens] += 1
        else:
            numTokens+=1
print "num2:%d" %  num2       
print "numTokens:%s" % numTokens
print numTerms

for t in terms:
    print t, terms[t]
```
