```python
# statistics for all problems 

import sqlite3
from stemming.porter2 import stem
from nltk.stem.snowball import SnowballStemmer
from nltk.stem.lancaster import LancasterStemmer
from nltk.stem import WordNetLemmatizer 


lan_stem = LancasterStemmer()
snow_stem = SnowballStemmer('english')
word_lemma = WordNetLemmatizer()




conn=sqlite3.connect("/Users/dc/clinical.sqlite")
c = conn.cursor()

c.execute("select distinct(problem_name) from problems;")
list_Problems = c.fetchall()
print len(list_Problems)


stopwords=set(['i','of','in','and','for','with','on','to','or','not','as','the'])
word_set=set()
for problems in list_Problems:
    #print "processing problem:", problems
    for words in problems[0].split():
        w=words.lower().strip(',').strip('(').strip(')').strip()
        
        if set([w]).issubset(stopwords)==False:
            #print "     adding word:", w
            word_set.add(w)
print "numwords:", len(word_set)
#print word_set
print '-------------------------------'
while (len(word_set)>1):
    word = word_set.pop()
    for test_stem in word_set:
        if word==lan_stem.stem(test_stem):
            print 'lancaster stemmer equal:', word, ' lan_stem:',test_stem
        elif word==snow_stem.stem(test_stem):
            print 'snowball stemmer equal:', word, 'snow_stem:', test_stem
        elif word==word_lemma.lemmatize(test_stem):
            print "lemma equal:", word, " to:", test_stem
#test stemming 
# 1 to n; n = 2402
print 'asdfasdfasdf'

```
