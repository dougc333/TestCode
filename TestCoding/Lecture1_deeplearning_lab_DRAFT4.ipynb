# Optimization and gradient descent

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

import numpy as np
import pylab as pl

def f(w):
    return w[0]**2 + 2*w[1]**2

def gradient(w):
    return np.array([2*w[0],4*w[1]])

w = np.array([-5,6])
z = f(w)

history = []
history.append([w,z])

#eta=0.01   # too small!
eta=0.1    # OK
#eta=0.5    # too big!

ITERS = 10  

for iter in range(ITERS):
    w = w - eta * gradient(w)
    z = f(w)
    history.append([w,z])
    

def build_mesh(lim,inc,f):
    X = np.arange(-lim, lim, inc)
    Y = np.arange(-lim, lim, inc)
    X, Y = np.meshgrid(X, Y)
    rowRange = range(X.shape[0])
    colRange = range(X.shape[1])
    Z= np.zeros((X.shape[0],X.shape[1]))
    for i in rowRange:
        for j in colRange:
            Z[i,j] = f(np.array([X[i,j],Y[i,j]]))    
    return X,Y,Z

X,Y,Z = build_mesh(8,0.25,f)  
fig = plt.figure(figsize=(20,14))
ax = fig.add_subplot(111, projection='3d')

xlist = np.array(map(lambda val: val[0][0],history))
ylist = np.array(map(lambda val: val[0][1],history))
zlist = np.array(map(lambda val: val[1],history))

jitter = True
if jitter:
    noise = 0.2
    xlist = xlist + np.random.normal(0, noise, size=len(xlist))
    ylist = ylist + np.random.normal(0, noise, size=len(ylist))
    zlist = zlist + np.random.normal(0, noise, size=len(zlist))

fig = plt.figure(figsize=(20,14))
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, 
        linewidth=0, antialiased=False,cmap=cm.Blues,alpha=0.3)
ax.contour(X,Y,Z,zdir='z',offset=-2)

ax.plot(xlist,ylist,zlist, color='r', marker='o', label='Gradient decent')
#plt.savefig('/home/charles/testbed_course/gd_surface_toobigl.png')

plt.show()

fig = plt.figure(figsize=(20,14))
cs = pl.contour(X,Y,Z,10,zdir='z',offset=-2)
pl.clabel(cs,inline=1,fontsize=10)
pl.plot(xlist, ylist, color='r', marker='o',alpha=0.5)
pl.plot(xlist[0], ylist[0], 'go', alpha=0.5, markersize=10)
pl.text(xlist[0], ylist[0], '  start', va='center')
pl.plot(xlist[-1], ylist[-1], 'ro', alpha=0.5, markersize=10)
pl.text(xlist[-1], ylist[-1], '  stop', va='center')
#plt.savefig('/home/charles/testbed_course/gd_contour_toobig.png')
plt.show()
```

#### some utility functions we will be using for the remainder of the tutorial

```python
import pylab as pl

from sklearn.datasets import make_moons, make_circles, make_classification
from matplotlib.colors import ListedColormap

def logistic(t):
    idx = t > 0
    out = np.copy(t)
    out[idx] = 1. / (1 + np.exp(-t[idx]))
    exp_t = np.exp(t[~idx])
    out[~idx] = exp_t / (1. + exp_t)
    return out

def softmax(x):
    if (len(x.shape)==1):
        x=x.reshape(1,-1)
    x -= np.max(x,axis=1).reshape(-1,1) 
    expX=np.exp(x)
    return expX/np.sum(expX,axis=1).reshape((-1,1))

def add_bias(X):
    return np.hstack((np.ones((X.shape[0],1)),X))

def one_hot(y):
    if y.ndim == 1:
        y = y.reshape((-1,1))    
    return (y == np.unique(y)).astype(int)


def plot_classify(X,y,predict,cm= pl.cm.Spectral,use_prob=False,scatter_cm=pl.cm.Spectral):
    step = .02  # step size in the mesh
    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, step),np.arange(y_min, y_max, step))
    m, n = np.c_[xx.ravel(), yy.ravel()].shape
    mesh = np.c_[xx.ravel(), yy.ravel()]
    if (use_prob):
        Z = predict(mesh)[1][:,1]
    else:
        Z = predict(mesh)[0]
    
    Z = Z.reshape(xx.shape)
    cm_bright = ListedColormap(['red','yellow','blue' ])
    fig = plt.figure(figsize=(12,12))
    ax = fig.add_subplot(1,1,1)
    markerSize = 40

    ax.contourf(xx, yy, Z, cmap=cm, alpha=.9)
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap=scatter_cm,s = markerSize)
    plt.show()
    
```

Linear Neuron
==============

Linear neuron allows us to learn linear combinations of the inputs. It provides a building block both for regression as well as classification models. By itself, the linear neuron provides a continuos output signal. This serves as the basis for linear regression models. In this case the data is represented by a matrix of $\mathbf{X}$ of feature vectors $\mathbf{x_n}$. Along with this we are given a set targets $\mathbf{X}$

* $\mathbf{X} \in \mathbb{R}^{N \times D}$      - the data matrix of  _feature vectors_
* $\mathbf{y} \in \mathbb{R}^N$                     - data vector of _target values_
* $\mathbf{w} \in \mathbb{R}^{D}$               - vector of _weights_ we want to learn

<img src="files/images/linear.png" width=900 height=600>

### $$\mathbf{f}(\mathbf{x}) =   \mathbf{w}^T\mathbf{x} = \sum_{d=1}^D w_d{x}_d = w_1{x}_1 + w_2{x}_2 + \ldots + w_D{x}_D$$

### $$loss(\mathbf{w},\mathbf{X}, \mathbf{y}) =\frac{1}{2N}(\mathbf{X}\mathbf{w} - \mathbf{y})^2 = \frac{1}{2N}\sum_{i=1}^N(\mathbf{w}^T\mathbf{x}_{i} - {y}_{i})^2 = mean(error)^2 = MSE$$ 

### $$\bigtriangledown_\mathbf{w} loss(\mathbf{w},\mathbf{X}, \mathbf{y})  = \frac{1}{N}\sum_{i=1}^N(\mathbf{w}^T\mathbf{x}_{i} - {y}_{i})\mathbf{x}_{i}  = \frac{1}{N}\sum_{i=1}^N\delta_{i}\mathbf{x}_{i} = \frac{1}{N}\sum_{i=1}^Nerror_{i}\mathbf{x}_{i}$$

###     $\mathbf{w}^{i + 1} = \mathbf{w}^i - \eta  \bigtriangledown loss_w$

```python
# generate  some data
slope=2.3
intercept=-3
N =300
sigma =10
X_raw = np.linspace(0,2*np.pi,N)
y_raw = slope*X_raw + intercept + sigma*numpy.random.normal(0, 0.2, N) 

X_raw= X_raw.reshape((-1,1))
y= y_raw.reshape((-1,1))

fig = plt.figure(figsize=(12,12))

plt.plot(X_raw, y, 'bo',alpha=0.6)
plt.show()

X = add_bias(X_raw)
y=y_raw
N,D = X_raw.shape
```

```python
step=0.001

MAX_ITERS= 10000
PRINT_ITER = 1000

W2 = np.random.uniform(-0.01,0.01,size=(D+1)) # init weights

C = 0                                         # regularization parameter


for iters in range(MAX_ITERS):    
    
    # forward : fprop
    Xw = X.dot(W2)       
    yHat= Xw   
     
    # backward: backprop
    error2 = (y - yHat)/N
    delta2 = error2
    dW2    = -(X.T).dot( delta2 )  + C*W2
    
    # print progress
    if (iters%PRINT_ITER == 0):
        cost = (y-yHat).dot(y-yHat) + 0.5*C*(W2.dot(W2))
        print iters,cost        
    
    # update weights
    W2 -= step * dW2
    
print W2

fig = plt.figure(figsize=(12,12))
plot(X_raw,yHat)
plt.plot(X_raw, y, 'bo',alpha=0.5)
plt.plot(X_raw,yHat, 'r--')
plt.show()   
```

# logistic neuron

## $$\sigma(x) = \frac{1}{1+e^{-\,x}}$$

```python
def sigmoid(x): 
    return 1.0/(1.0+exp(-x))

x = linspace(-8,8,100)
fig = plt.figure(figsize=(14,10))
plot(x,sigmoid(x))
```

### 2 classes:   $ C_1, C_2$ <br/> <br/>   $$p(C_1|\mathbf{x}) = f(a) = \sigma(\mathbf{w}^T\mathbf{x})$$  $$p(C_2|\mathbf{x}) = 1- p(C_1|\mathbf{x})  = 1 - \sigma(\mathbf{w}^T\mathbf{x})$$   <br>  $$likelihood(\mathbf{y}|\mathbf{w},\mathbf{x}) = \prod_{n=1}^N p(C_1|\mathbf{x})^{y_n} (1 - p(C_1|\mathbf{x}))^{1-y_n}$$   $$ loss(\mathbf{w},\mathbf{X}, \mathbf{y}) = -negative log likelihood(Data) $$  $$ loss(\mathbf{w},\mathbf{X}, \mathbf{y}) = - \frac{1}{N}\sum^N_{n=1}[\mathbf{y}_n \log           p(C_1|\mathbf{x}) + (1 - \mathbf{y}_n)\log(1 - p(C_1|\mathbf{x}))] $$ <br> $$\bigtriangledown_\mathbf{w} loss(\mathbf{w},\mathbf{X}, \mathbf{y}) = \frac{1}{N}\sum_{n=1}^N (\sigma(\mathbf{w}^T \mathbf{x}_n)) - y_n) \mathbf{x}_n= \frac{1}{N}\sum_{n=1}^N\delta_{n}\mathbf{x}_{n} = \frac{1}{N}\sum_{n=1}^Nerror_{n}\mathbf{x}_{n}$$ 

###     $\mathbf{w}^{i + 1} = \mathbf{w}^i - \eta  \bigtriangledown loss_w$ 

### generate data

```python
# generate X
N = 600
M = 2
X_raw = np.random.randn(N,M)
X = add_bias(X_raw)

w_true = np.array([0.0,2.0,1.4])
probs =  logistic(X.dot(w_true))
draws =  np.random.uniform(0,1,N)

# generate labels
y_raw = np.array([1.0 if draw < prob else 0.0 for (draw,prob) in zip(draws,probs)])

cm_bright = ListedColormap(['#FF0000', '#0000FF'])

fig = plt.figure(figsize=(12,12))
plt.scatter(X_raw[:, 0], X_raw[:, 1], c=y_raw, cmap=cm_bright)
plt.show()

y   = y_raw
N,D = X_raw.shape
K   = len(np.unique(y_raw))
```

```python
step=0.001

MAX_ITERS= 10000
PRINT_ITER = 1000

W2 = np.random.uniform(-0.01,0.01,size=(D+1))  # init weights

C = 1                                          # regularization param

for iters in range(MAX_ITERS):    
    
    # forward: fprop
    Xw = X.dot(W2)       
    yHat= logistic(Xw)    
    
    # backward: backprop
    error2 = (y - yHat)
    delta2 = error2
    dW2    = -(X.T).dot( delta2 )  + C*W2
    
    # print progress
    if (iters%PRINT_ITER == 0):
        #w = W2.flatten()
        cost = ( y.dot(np.logaddexp(0,-Xw))  + (1.0-y).dot(np.logaddexp(0,Xw))) + 0.5*C*(W2.dot(W2))
        print iters,cost  
        
    # update weights
    W2 -= step * dW2        
    
def predict_logistic(W2,X):
    X = add_bias(X)    
    Xw = np.dot( X, W2 )       
    probs = logistic(Xw)
    pred = np.array([1.0 if p > 0.5 else 0.0 for p in probs])
    prob_array = np.array([ [1.0 - p,p] for p in probs])
    return (pred,prob_array)

predict  = lambda X  : predict_logistic(W2,X) 
plot_classify(X_raw,y_raw,predict,use_prob=False,cm=pl.cm.RdBu,scatter_cm=cm_bright)    


```

# Softmax neuron 

### K classes:   $ C_1, C_2, ..., C_K$ <br/> <br/>   $$f = p(C_k | \mathbf{x}) = \frac{\exp(a_k)}{\sum_j \exp(a_j)}$$ <br/>   $$a_j = \ln(p(\mathbf{x}|C_k)p(C_k))$$<br/>   $$ loss(\mathbf{w},\mathbf{X}, \mathbf{y}) = - \sum_{n=1}^N \sum_{k=1}^K y_{nk} \ln p(C_k | \mathbf{x})$$ <br>  $$\bigtriangledown_\mathbf{w_k} loss(\mathbf{w},\mathbf{X}, \mathbf{y}) = \frac{1}{N}\sum_{n=1}^N ( p(C_k | \mathbf{x})  - y_nk) \mathbf{x}_n = \frac{1}{N}\sum_{n=1}^N\delta_{n}\mathbf{x}_{n} = \frac{1}{N}\sum_{n=1}^Nerror_{n}\mathbf{x}_{n}$$

```python
step=0.001
MAX_ITERS= 10000
PRINT_ITER = 1000

y = one_hot(y_raw)                              # softmax needs one hot encoding

W2 = np.random.uniform(-0.01,0.01,size=(D+1,K))  # init weights

C = 1                                            # regularization param

for iters in range(MAX_ITERS):    
    
    # forward : fprop
    Y = X.dot(W2)       
    yHat= softmax(Y)   
     
    # backward: backprop
    error2 = (y - yHat)
    delta2 = error2
    dW2    = -(X.T).dot( delta2 )  + C*W2
    
    # print progress
    if (iters%PRINT_ITER == 0):
        w = W2.flatten()
        cost= -np.sum(y * np.log(yHat)) + 0.5*C*(w.dot(w))
        print iters,cost        
    
    # update weights
    W2 -= step * dW2
    
def predict_softmax(W2,X):
    X = add_bias(X)    
    Y = np.dot( X, W2 )       
    Y = softmax(Y)
    return (np.argmax(Y,axis=1),Y)

predict  = lambda X  : predict_softmax(W2,X) 
plot_classify(X_raw,y_raw,predict,use_prob=True,cm=pl.cm.RdBu,scatter_cm=cm_bright)   
  
```

### Single hidden layer nnet

```python
step=0.001
MAX_ITERS= 10000
PRINT_ITER = 1000

H =10                                             # number of hidden nearons

W1= np.random.uniform(-0.01,0.01,size=(D+1,H))    # init input to hidden weights
W2 = np.random.uniform(-0.01,0.01,size=(H+1,K))   # init hidden to output weights

C = .01                                           # regularization param

for iters in range(MAX_ITERS):    
    
    # forward : fprop
    Z = logistic( X.dot(W1) )
    Z1 = add_bias(Z)
    Y = Z1.dot(W2)       
    yHat= softmax(Y)   
    
    # backward: backprop
    error2 = (y - yHat)
    delta2 = error2
    dW2    = -(Z1.T).dot( delta2 )  + C*W2
    
    error1 = delta2.dot(W2[1:,:].T)
    delta1 = error1 * Z * (1.0 - Z)
    dW1    = -(X.T).dot(delta1) + C*W1  
    
    # print progress
    if (iters%PRINT_ITER == 0):
        w = np.hstack((W1.flat,W2.flat))
        cost= -np.sum(y * np.log(yHat)) + 0.5*C*(w.dot(w))
        print iters,cost 
        
    # update weights    
    W2 -= step * dW2
    W1 -= step * dW1
    
def predict_nnet(W1,W2,X):
    X = add_bias(X)    
    Z = logistic(np.dot( X, W1 ))
    Z1 = add_bias(Z)
    Y = np.dot( Z1, W2 )       
    Y = softmax(Y)
    return (np.argmax(Y,axis=1),Y)

predict  = lambda X  : predict_nnet(W1,W2,X) 
plot_classify(X_raw,y_raw,predict,use_prob=False)    
```

# Using scipy optimize routines

```python
import math
import pylab as pl
import scipy
from numpy.linalg import norm as norm
from matplotlib.colors import ListedColormap
from scipy.optimize import fmin_bfgs

def forward(w,X,N,H,K):
    V = w[:(N+1)*H].reshape((N+1,H))
    W = w[(N+1)*H:].reshape((H+1,K)) 
    Z = logistic(np.dot( X, V ))
    Z1 = add_bias(Z)
    Y = np.dot( Z1, W )       
    Y = np.dot(add_bias(logistic(np.dot(X,V))), W)   
    return (V,W,Z,softmax(Y))

def nnet_cost(w,X,y,C,N,H,K):
    (V,W,Z,yHat) = forward(w,X,N,H,K)
    return -np.sum(y * np.log(yHat)) + 0.5*C*w.dot(w)

def grad_nnet_cost(w,X,y,C,N,H,K):
    (V,W,Z,yHat) = forward(w,X,N,H,K)
    error = (y - yHat) 
    Z1 = add_bias(Z)      
    dW = -np.dot( Z1.T, error)  + C*W
    dV = -np.dot( X.T, np.dot( error, W[1:,:].T) * Z*(1.0-Z)) + C*V
    return np.hstack((dV.flat,dW.flat))

def predict_nnet(weights,X,N,H,K):
    X = add_bias(X)
    (V,W,Z,Y) = forward(weights,X,N,H,K)
    return (np.argmax(Y,axis=1),Y)

def train(w0,X,y,cost,gradient):         
    weights = fmin_bfgs(cost,x0 = w0, fprime = gradient,args = (X, y),gtol = 1e-6)
    return weights
```
